```python
import pandas as pd
import numpy as np
from padronizar_data import padronizar_periodo
```

```python
#Carregamento e amostragem das primeiras linhas da dataframe


df_alunos = pd.read_csv(r"C:\Users\blood\Documents\bolsistas_2025.csv",low_memory=False)
display(df_alunos.head())
```

```python
#Descrição estatística da dataframe 


display(df_alunos.describe())
```

```python
# Tratamento de dados
#Criando uma cópia da dataframe e fazendo as modificações necessárias para facilitar a manipulação dos dados 

df_analise = df_alunos
df_analise = df_analise.drop(columns=['Numero'])
df_analise = df_analise.dropna()
df_analise['Curso'] = df_analise['Curso'].str.strip()
df_analise['Curso'] = df_analise['Curso'].str.title()
df_analise['Valor mensal da bolsa'] = df_analise['Valor mensal da bolsa'].str.strip('R$')
df_analise['Valor mensal da bolsa'] = df_analise['Valor mensal da bolsa'].str.strip('')
df_analise['Valor mensal da bolsa'] = df_analise['Valor mensal da bolsa'].str.replace(',','.')
df_analise['Valor mensal da bolsa'] = df_analise['Valor mensal da bolsa'].astype(float)
df_analise['Período de recebimento da bolsa'] = df_analise['Período de recebimento da bolsa'].str.strip('')
df_analise['Período de recebimento da bolsa'] = df_analise['Período de recebimento da bolsa'].apply(padronizar_periodo)


display(df_analise)
display(df_analise.info())
```

```python
# Analisando dados de acordo com filtros específicos
# Alunos com bolsa maior ou igual a 700 reais 
# Alunos bolsistas de cursos de licenciatura 

display(df_analise[df_analise['Valor mensal da bolsa'] >= 700])
display(df_analise[df_analise['Curso'].str.contains('Licenciatura',na=False)])
```
